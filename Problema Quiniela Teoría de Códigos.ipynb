**Lo primero que vamos a necesitar es un método que, introducidos dos códigos, nos diga si uno recubre al otro:**
    
    El método implementado devuelve el número de palabras que han sido recubiertas, es decir, si queremos comprobar si un código de 10 palabras ha sido totalmente recubierto por el segundo código introducido, esperaremos una respuesta del valor 10 por parte del método.

```python
def getRecubiertos (codigoCompleto, recubridor):
    ct = 0;
    #Estos 2 primeros bucles hacen que se recorrar el código recubridor una vez por cada palabra del código a comprobar.
    for i in codigoCompleto:
        for k in recubridor:
            d = 0;
            #En este bucle se recorren cada uno de los dígitos en busca de diferencias.
            for j in range(6):
                if(i[j] != k[j]):
                    d = d+1;
            #Si d < 2 la distancia se respeta y la palabra es recubierta, por tanto se suma uno al valor de respuesta.
            if(d<2):
                ct = ct+1;
                break
    return ct
```

**A continuación vamos a crear dos métodos necesarios. El primero "generateD" que generará una palabra d a partir de una palabra pasada por parámetro. Esta palabra d estará a una distancia 1 mediante el cambio aleatorio de uno de los dígitos de la palabra original.**

**El segundo método "aceptaDporTemp" nos servirá para conocer el resultado probabilístico acertado o fallido según el cual nos quedaremos o no con el valor de d en el código de recubrimiento en cierta parte del algoritmo.**

```python
import random, math, copy

def generateD (word):
    #Seleccionamos un dígito aleatorio
    elem = randrange(6)
    newWord = copy.deepcopy(word)
    #Cambiamos el valor del dígito a un valor distinto respecto al valor original
    if (newWord[elem]==0):
        newValue = random.choice([1,2]);
    if (newWord[elem]==1):
        newValue = random.choice([0,2]);
    if (newWord[elem]==2):
        newValue = random.choice([0,1]);
    newWord[elem] = newValue  
    
    #Por último devolvemos la palabra generada
    return newWord


def aceptaDporTemp(N, T):
    result = False
    #Generamos un número aleatorio de 0 a 1
    a = random.uniform(0, 1)
    #Calculamos la probabilidad según N y T
    b = math.exp(-N/T)
    #Si el número aleatorio está dentro de la probabilidad se devuelve True, de lo contrario será False
    if(a<=b):
        result = True
    return result
```

**Ya disponemos de todos los elemento necesarios para llevar a cabo la tarea. Veamos cómo implementar el algoritmo del enfriamiento del acero:**

```python
#S -> las 270 apuestas del Código reducido.
#n -> número de palabras del código de recubrimiento.
#C -> Código de recubrimiento.
#N -> Número de apuestas en S que aún no se recubren con C, que tienen distancia > 1 a cualquier palabra de C.
#d -> Palabra a distancia 1 del ci actual.

#El código base desde el que construiremos el nuevo código recubridor lo generaremos cogiendo el número de palabras que deseemos
#tener finalmente del código recubridor aportado en el enunciado aunque no tendría necesariamente que ser así, lo llamamos C.
aux = recubridor
#Para seleccionar el tamaño deseado del código de recubrimiento podemos modificar el valor 50 a continuación. 
#Se han llegado a encontrar buenos resultados para TAM=44 pero se tarda una cantidad de tiempo bastante grande.
C = aux[0:50]
#Para saber nuestro N cogeremos el tamaño del código a reducir y le restaremos el número de palabras recubiertas por el
#código recubridor. Si el resultado es 0, el código ha sido completamente recubierto.
N = len(reduced) - getRecubiertos(reduced, C)
#T Temperatura, F Frío.
T = 2
F = 0.95

import time

start = time.time()

while N>0: 
    print("... N actual: ", N)
    #Recorremos toda la instancia actual del código recubridor.
    for i in C:
        ind = C.index(i)
        #Repetimos el siguiente proceso para cada palabra un número determinado de veces.
        for j in range(5):
            #Generamos una palabra d con nuestro método.
            ci = i
            d = generateD(i)
            i = d
            C[ind] = d
            #Se ha invertido un poco la lógica del algoritmo. En esta ocasión añadimos d por defecto al código y si se
            #incumplen las considicones revertimos ese cambio.
            if(N<(len(reduced) - getRecubiertos(reduced, C))):
                #Si no se cumple la probabilidad revertimos el cambio mediante nuestro método.
                if(aceptaDporTemp(N, T) == False):
                    i = ci
                    C[ind] = ci
            else:
                #Actualizamos N y T, si N ha llegado a 0 (Objetivo del algoritmo) interrumpimos el bucle.
                N = len(reduced) - getRecubiertos(reduced, C)
            if((len(reduced) - getRecubiertos(reduced, C)) == 0):
                print("Código recubridor: ", C)
                break        
        T = T*F
        if((len(reduced) - getRecubiertos(reduced, C)) == 0):
            break 
            
end = time.time()
print(end - start)
    
```
